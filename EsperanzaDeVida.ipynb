```python
#Importar librerías
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importamos los datos
df = pd.read_csv("Info_pais.csv", encoding= "ISO-8859-1", delimiter=";")
df_ordenado = df.sort_values("Esperanza de vida", ascending=True)
df.head()
```

```python
#Visualización de la gráfica Primer eje es X, segundo eje es y
plt.scatter(df_ordenado["Renta per capita"], df_ordenado["Esperanza de vida"])

#Agregar un título a la gráfica
plt.title("Renta per capita vs Esperanza de vida")

#Etiqueta a X
plt.xlabel("Renta per capita")

#Etiqueta a y
plt.ylabel("Esperanza de vida")
```

```python
#Crear nueva columna
df_ordenado["Población normalizada"] = df_ordenado["Poblacion"]/(max(df_ordenado["Poblacion"])/10000)
```

```python
df_ordenado.head()
```

```python
#Visualización de la gráfica Primer eje es X, segundo eje es y                  s = size la variable y el contenido a mostrar   c = color
plt.scatter(df_ordenado["Renta per capita"], df_ordenado["Esperanza de vida"], s=df_ordenado["Población normalizada"], c=df_ordenado["Población normalizada"])

#Agregar un título a la gráfica
plt.title("Renta per capita vs Esperanza de vida")

#Etiqueta a X
plt.xlabel("Renta per capita")

#Etiqueta a y
plt.ylabel("Esperanza de vida")

#Tamaño del gráfico
fig = plt.gcf()
fig.set_size_inches(12, 10)

#Añadir la etiqueta del país a los circulos
for i in range(1, 10):
    plt.annotate(df_ordenado["País"][i], (df_ordenado["Renta per capita"][i], df_ordenado["Esperanza de vida"][i]))

#Acomodar los datos amontonados
plt.yticks(ticks=range(1,120,10))
```
